```python
df=pd.read_csv('/content/example_data.csv')
```

```python
weird=(df['점수']>100) | (df['공부시간']>10)
weird_data=df[weird]
print(len(weird))
display(weird_data)
```

```python
df_no_duplicates = df.drop_duplicates()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.scatter(df['점수'],df['공부시간'])
plt.xlabel('공부시간')
plt.xlabel('점수')
```

```python
df['공부시간'].quantile(0.25)
df['공부시간'].quantile(0.75)
IQR = Q3 - Q1
lowtime=Q1 - 1.5 * IQR
hightime=Q3 + 1.5 * IQR
```

```python
df['점수'].quantile(0.25)
df['점수'].quantile(0.75)
IQR = Q3 - Q1
lowscore=Q1 - 1.5 * IQR
highscore=Q3 + 1.5 * IQR
```

```python
outtime=df[(df['공부시간']<lowtime)] | (df['공부시간']>hightime)
outtime_data=df[outtime]
print(outtime)
display(outtime)
```

```python
outscore=df[(df['점수']<lowscore)] | (df['점수']>highscore)
outscore_data=df[outscore]
print(outscore)
display(outscore)
```

```python
df_no_out = df[
    (df['공부시간'] >= lowtime) & (df['공부시간'] <= hightime) &
    (df['점수'] >= lowscore) & (df['점수'] <= highscore)
]
df_no_out=df_no_out.reset_index(drop=True)
display(df_no_out)
```

```python
plt.scatter(df_no_out['점수'],df_no_out['공부시간'])
plt.xlabel('공부시간')
plt.xlabel('점수')
```

```python
class_mean=df_no_out.groupby('분반')['점수'].mean()
plt.bar(class_mean.index, class_mean.values)
plt.xlabel('분반')
plt.ylabel('평균 점수')
plt.title('각 분반의 평균 점수')
plt.show()
```
